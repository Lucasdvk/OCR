```python
## Importação das bibliotecas

import pytesseract as ts
import numpy as np
import cv2 
from matplotlib import pyplot as plt
```

```python
# Leitura de imagem

img = cv2.imread('imagem.jpg')

# Converção de BGR para RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Exibir a imagem usando Matplotlib
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
# Converção para texto

texto = ts.image_to_string(img_rgb)
print (texto)
```
