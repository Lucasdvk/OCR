```python
import cv2
import pytesseract as ts
import pandas as pd
import matplotlib.pyplot as plt

# Carregar a imagem
img = cv2.imread('exp.jpg')

# Converter a imagem para RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Visualizar a imagem
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
# Usando o Tesseract para obter o texto da imagem
config_tesseract = '--tessdata-dir tessdata --psm 6 --oem 1'
texto = ts.image_to_string(img_rgb, lang='por', config=config_tesseract)

print(texto)
```
